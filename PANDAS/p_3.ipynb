✅ 5. Simple Stock Price Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = ["Day 1","Day 2","Day 3","Day 4","Day 5","Day 6","Day 7","Day 8","Day 9","Day 10"]
prices = [102, 105, 101, 108, 110, 107, 112, 115, 113, 118]

plt.figure(figsize=(6, 6))
plt.plot(days,prices,marker="o",color="red",linestyle="-")
plt.fill_between(days,prices,color='pink')
plt.grid(color="blue",linestyle="-")
plt.scatter(days, prices, color='black', zorder=3)
plt.title("Stock Price Trend", fontsize=16, fontweight='bold')
plt.xlabel("Days", fontsize=12, fontweight='bold')
plt.ylabel("Price ($)", fontsize=12, fontweight='bold')
plt.xticks(rotation=45,fontweight='bold',fontsize=18,color="blue")
plt.yticks(fontweight='bold',fontsize=18 , color="blue")
for i in range(1,len(days)):
    if prices[i] > prices[i-1]:
        plt.plot(days[i-1:i+1],prices[i-1:i+1], color="green",linewidth=5)
    else:
        plt.plot(days[i-1:i+1],prices[i-1:i+1],color='red',linewidth=5)
plt.show()
```

✅ 1. COVID-19 Cases Trend Analysis

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
days = list(range(1, 31))  # Days 1 to 30
cases = [120, 150, 180, 200, 250, 300, 280, 320, 350, 400, 
         420, 450, 470, 490, 500, 520, 530, 540, 550, 560,
         570, 580, 590, 600, 610, 620, 630, 640, 650, 660]

# 7-day rolling average
rolling_avg = np.convolve(cases, np.ones(7)/7, mode='valid')

# Plot
plt.figure(figsize=(12, 6))
plt.plot(days, cases, color='blue', marker='o', label="Daily Cases")

# Add rolling average line (shift by 3 to center the avg window)
plt.plot(days[3:-3], rolling_avg, color='red', linewidth=2, label="7-Day Avg")

# Fill between actual and average
plt.fill_between(days[3:-3], rolling_avg, cases[3:-3], 
                 where=(cases[3:-3] > rolling_avg), color='green', alpha=0.2,
                 label="Above Avg")
plt.fill_between(days[3:-3], rolling_avg, cases[3:-3], 
                 where=(cases[3:-3] < rolling_avg), color='orange', alpha=0.2,
                 label="Below Avg")

# Highlight peak day
peak_day = days[cases.index(max(cases))]
peak_cases = max(cases)
plt.scatter(peak_day, peak_cases, color='purple', s=100, zorder=5)
plt.annotate(f'Peak: {peak_cases}', xy=(peak_day, peak_cases), xytext=(peak_day-5, peak_cases+50),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Labels and title
plt.title("COVID-19 Daily Cases Trend", fontsize=16, fontweight="bold")
plt.xlabel("Days", fontweight="bold")
plt.ylabel("Number of Cases", fontweight="bold")
plt.grid(alpha=0.3)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
rainfall_2023 = [80, 60, 75, 100, 120, 150, 200, 220, 180, 140, 90, 70]
rainfall_2024 = [70, 65, 80, 110, 130, 160, 210, 230, 190, 150, 95, 75]

x = np.arange(len(months))  # positions for bars

# Plot setup
plt.figure(figsize=(12, 6))

# Bar chart for 2023
bars = plt.bar(x, rainfall_2023, color='skyblue', alpha=0.8, label='2023')

# Line chart for 2024
plt.plot(x, rainfall_2024, color='orange', marker='o', linewidth=2, label='2024')

# Highlight wettest month in 2024
wettest_idx = np.argmax(rainfall_2024)
wettest_month = months[wettest_idx]
wettest_value = rainfall_2024[wettest_idx]

plt.scatter(wettest_idx, wettest_value, color='red', s=100, zorder=5)
plt.annotate(f'Wettest: {wettest_month}\n{wettest_value} mm',
             xy=(wettest_idx, wettest_value), xytext=(wettest_idx+0.5, wettest_value+20),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10, fontweight='bold')

# Highlight highest bar (2023)
max_bar_idx = np.argmax(rainfall_2023)
bars[max_bar_idx].set_color('blue')

# Customize
plt.xticks(x, months)
plt.title("Monthly Rainfall Comparison (2023 vs 2024)", fontsize=16, fontweight='bold')
plt.xlabel("Months", fontsize=12, fontweight='bold')
plt.ylabel("Rainfall (mm)", fontsize=12, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
