# series


```python
import pandas as pd
series_1 = pd.Series([10,20,30,40],index =['a','b','c','d'])
series_2 = pd.Series(["reddy","mani","kamal","pawan"],index=['a','b','c','d'])
result = pd.concat([series_1,series_2],axis=1)
result.columns=["age","marks"]
print(result)
print()
print("-----------------------SERIES OPERATIONS------------------------------")
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(s1 + s2)   # Addition
print(s1 - s2)   # Subtraction
print(s1 * s2)   # Multiplication
print(s1 / s2)   # Division
print()
print()
print("--------2. Statistical Operations----------")
print("Sum:", s1.sum())
print("Mean:", s1.mean())
print("Max:", s1.max())
print("Min:", s1.min())
print("Standard Deviation:", s1.std())
print()
print()
print("-----------------3. Index-based Selection---------")
print(s1['a'])   # Get by label
print(s1[0])     # Get by position
print(s1[['a','c']])  # Multiple indices
print()
print(s1[s1 > 20])   # Filter values greater than 20
print(s1 % 2 == 0)   # Check even numbers
s3 = pd.Series(["Reddy", "Mani", "Kamal", "Pawan"])
print(s3.str.upper())   # Convert to uppercase
print(s3.str.len())     # Length of each string
print(s3.str.contains("a"))  # Check if 'a' is present
s4 = pd.Series([10, None, 30, None, 50])
print(s4.isnull())      # Check NaN
print(s4.notnull())     # Check not NaN
print(s4.fillna(0))     # Replace NaN with 0
print(s4.dropna())      # Remove NaN
print(s1.apply(lambda x: x**2))  # Square each value
s5 = pd.Series([100, 200, 300], index=['a', 'b', 'x'])
print(s1 + s5)   # Auto-aligns by index


```
