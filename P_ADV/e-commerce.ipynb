```python
import numpy as np
import pandas as pd
```

```python
customer_table = pd.read_csv("customers.csv")
oreder_table = pd.read_csv("orders.csv")
product_table = pd.read_csv("products.csv")

merge_table = pd.merge(customer_table,oreder_table, on="CustomerID", how ="right" )

m_3 = pd.merge(merge_table,product_table , on="ProductID",how="right")
# print(m_3)

m_3[["OrderDate","ShipDate"]] = (m_3[["OrderDate","ShipDate"]]).apply(pd.to_datetime)

# null = m_3.info()
# print(null)

fill_nill = m_3.fillna(1111)
print(fill_nill)
drop = m_3.dropna()
print(drop)
```

```python
print("----TOTAL SALES PROFIT QUANTITY BY REGION-----------")
region = m_3.groupby("Region")[['Sales','Profit','Category']].sum()
print(region)
print()
print("----TOTAL SALES PROFIT QUANTITY BY Segment-----------")
segment = m_3.groupby("Segment")[['Sales','Profit','Category']].sum()
print(segment)
print()
print("----TOTAL SALES PROFIT QUANTITY BY Category-----------")
category = m_3.groupby("Category")[['Sales','Profit','Category']].sum()
print(category)

```

```python
# m_3["Profit_margin"] = m_3["Profit"] / m_3["Sales"]
# print(m_3.head())
m_3["yearmonth"] = m_3["OrderDate"].dt.to_period("M")
m_s = m_3.groupby("yearmonth")["Sales"].sum()
print(m_s)
```

```python
print("----top 10 consumers-----")
top = m_3.groupby("CustomerName")[["Sales", "Profit"]].sum().sort_values(by="Sales",ascending=False).head(10)
print(top)
```

```python
print("----top 10 consumers-----")
top = m_3.groupby("CustomerName")[["Sales", "Profit"]].sum().sort_values(by="Profit",ascending=False).head(10)
print(top)
```

```python

m_3["year"] = m_3["OrderDate"].dt.year

yearly_sales = m_3.groupby("year")["Sales"].sum()
yoy_growth = yearly_sales.pct_change()*100
print(yoy_growth)
```

```python
m_3["Delivery_days"] = (m_3["ShipDate"] - m_3["OrderDate"]).dt.days

avg = m_3["Delivery_days"].mean()

delay = m_3[m_3["Delivery_days"]>5]

print("AVG :- ",avg)
print("Delayed Orders:", len(delay))
```
